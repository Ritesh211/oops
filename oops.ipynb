9. What is a constructor in Python?
>In Python, a constructor is a special method used to initialize a newly created object of a class. It is automatically called when an object of the class is instantiated.
Key characteristics of a Python constructor:
Name:
The constructor method in Python is always named __init__. This is a special method name recognized by Python.
Purpose:
Its primary purpose is to set up the initial state of the object. This involves assigning values to the object's attributes (instance variables) or performing any other setup logic required for the object to be ready for use.
Automatic Invocation:
Unlike regular methods that must be explicitly called, the __init__ method is automatically invoked by Python when an object of the class is created.
self parameter:
The first parameter of the __init__ method is always self. This parameter refers to the instance of the object being created, allowing you to access and modify its attributes.
Parameters:
Constructors can accept additional parameters beyond self to receive initial values for the object's attributes.

 17. What is the difference between a class variable and an instance variable
 > The fundamental difference between a class variable and an instance variable lies in their scope, storage, and how they are accessed and shared among objects of a class.
Class Variables:
Definition:
Class variables are declared directly within the class definition, but outside of any methods. In some languages like Java, they are often declared with the static keyword.
Scope & Sharing:
They belong to the class itself and are shared by all instances (objects) of that class. There is only one copy of a class variable, regardless of how many instances are created.
Access:
Class variables can be accessed using the class name directly (e.g., ClassName.variable_name) or through an instance of the class.
Use Cases:
Ideal for storing data that is common to all instances, such as constants, configuration settings, or counters that track the number of instances created.
Instance Variables:
Definition:
Instance variables are declared within a class but are associated with individual instances of that class. In Python, they are typically defined within the __init__ method using self.variable_name.
Scope & Sharing:
Each instance of a class has its own unique copy of instance variables. Changes to an instance variable in one object do not affect the same variable in other objects.
Access:
Instance variables are accessed through an instance of the class (e.g., object_name.variable_name).
Use Cases:
Used to store data that is unique to each specific object, representing its individual state or characteristics.
In summary: Class variables are shared across all instances, while instance variables are unique to each instance. This distinction is crucial for designing object-oriented programs where data needs to be either common to all objects or specific to individual objects.

 22. What is the difference between @staticmethod and @classmethod in Python
 > @staticmethod
No implicit first argument:
A static method does not receive the instance (self) or the class (cls) as its first argument. It behaves like a regular function that is logically grouped within a class.
No access to class or instance state:
It cannot directly access or modify class-level attributes or instance-level attributes.
Use case:

Implicit first argument cls: A class method receives the class itself (cls) as its first argument, allowing it to access and modify class-level attributes.

23. How does polymorphism work in Python with inheritance
> Polymorphism in Python, particularly in conjunction with inheritance, allows objects of different classes to be treated as objects of a common type. This is achieved primarily through method overriding.
Here's how it works:
Inheritance:
A parent class (superclass) defines a method. Child classes (subclasses) inherit this method.
Method Overriding:
A child class can provide its own specific implementation for a method that it inherited from its parent class. This means the child class "overrides" the parent's method with its own version.
Dynamic Binding:
When you call a method on an object, Python determines which implementation of that method to use at runtime based on the actual type of the object, not just the type of the variable holding the object. If the object's class has overridden the method, that overridden version is executed. Otherwise, the inherited method from the parent class is used.
